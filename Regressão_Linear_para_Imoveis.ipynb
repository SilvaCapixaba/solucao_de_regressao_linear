```python
#aqui separei as variveis
X = base_casa[['Avg_AreaIcome', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms',
       'Avg_Area_Number_of_bedrooms', 'Area_population']] #variavel explicativa

Y = base_casa[["Price"]] #variavel resposta
```

```python
#importar uma biblioteca especifica para machinelearning
#aqui eu estou separando as base; estou criando duas base de teste e base de treino
#train_size=0.7 aqwui estou separadno a minha base que vai para treino são 70% da base - o normal da divisão e de 70 a 80 %
#random_state= 'aqui ele vai pegar os dados de modo aleatório
#X_train - Treino / X_test - Test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
#Aqui vou vizualizar a quantidade de dados que o modelo vai pegar para cada base - "mais por curiosidade"
print(X_train.shape) #base treino
print(X_test.shape) #base test
print(Y_train.shape)
print(Y_test.shape)
```

```python
#estou importando sklearn para regressão linear
from sklearn.linear_model import LinearRegression
```

```python
#estou colocando a LinearRegressão dentro de uma variavél
lm = LinearRegression()
```

```python
#agora estou treinando os meus mdelos de treino
lm.fit(X_train, Y_train)
```

```python
#agora vou tentar predizer este modelo a base de test - estou treinando para verificar mais adiante o modelo de test
Y_pred = lm.predict(X_test)
```

```python
#importando biblioteca para ter a metricas
#ENTENDENDO: QUANTO MAIOR O r2 SCORE, MAIOR E POSSIBILIDADE DE ASCERTO OU É MAIS PROX. DO OBJETIVO
from sklearn.metrics import r2_score #quanto maior r2 - muito bom esta meus dados e minhas analises vao ser mais ascertivas
r = r2_score(Y_test, Y_pred) #estou comparando o Y_test(eo Y da base)- Y_pred(é o Y da previsão do modelo) -a previsão do modelo para comparar com a qualidade dos meus dados-com relação as analises para possiveis ascertos
```

```python
#r2 - vaide -1 ate +1 -  meu r2 esta muito alto isso me da previsão de 915 de ascerto eu tenho uma linearidade muito alto no meu modelo
print("r_quadrado:", r)
```
